## Análise e plotagem dos sinais das simulações

### Importando os dados

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import scienceplots

%config InlineBackend.figure_format="retina"
plt.style.use(["science", "notebook"])

SMALL_SIZE = 10
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=SMALL_SIZE)        
plt.rc('axes', titlesize=BIGGER_SIZE)  
plt.rc('axes', labelsize=MEDIUM_SIZE)
plt.rc('xtick', labelsize=SMALL_SIZE)  
plt.rc('ytick', labelsize=SMALL_SIZE)  
plt.rc('legend', fontsize=SMALL_SIZE)  
plt.rc('figure', titlesize=BIGGER_SIZE)
```

```python
def graficos_rodadas(ensaiosV1, ensaiosV2, nome):
    print(f"\nMédia Rodadas - V1: {np.mean(ensaiosV1[0])}")
    print(f"Média Rodadas - V2: {np.mean(ensaiosV2[0])}\n")
    plt.figure(figsize=(8, 7))
    plt.subplot(2, 1, 1)
    plt.bar(range(len(ensaiosV1[0])), ensaiosV1[0], label="Rodadas - Agente V1", color="b")
    plt.bar(range(len(ensaiosV1[0])), ensaiosV2[0], linestyle='--',
             label="Rodadas - Agente V2", color="r")
    plt.xlabel("Ensaios")
    plt.ylabel("Rodadas")
    plt.grid(which='both')
    plt.grid(which='major', color='#CCCCCC', linestyle='-', alpha=1)
    plt.grid(which='minor', color='#CCCCCC', linestyle=':', alpha=0.5)
    plt.legend()
    plt.savefig(nome, format='svg')
    plt.show()
```

```python
def graficos_passos(ensaiosV1, ensaiosV2, nome):
    print(f"\nMédia Passos - V1: {np.mean(ensaiosV1[1])}")
    print(f"Média Passos - V2: {np.mean(ensaiosV2[1])}\n")
    
    plt.figure(figsize=(8, 7))

    plt.subplot(2, 1, 2)
    plt.bar(range(len(ensaiosV1[1])), ensaiosV1[1], label="Passos do Agente V1", color="b")
    plt.bar(range(len(ensaiosV2[1])), ensaiosV2[1], linestyle='--',
             label="Passos do Agente V2", color="r")
    plt.xlabel("Ensaios")
    plt.ylabel("passos")
    plt.grid(which='both')
    plt.grid(which='major', color='#CCCCCC', linestyle='-', alpha=1)
    plt.grid(which='minor', color='#CCCCCC', linestyle=':', alpha=0.5)
    plt.legend()
    plt.savefig(nome, format='svg')
    plt.show()
```

### Simulação 1 - Dimensão 3x3 Qt. Vitórias 50

```python
df_3x3 = pd.read_csv("../simulacao/sim_1_dimensao_3_v50", sep=";", header = None).values


graficos_rodadas(df_3x3[:2], df_3x3[2:], "img/rodadas_3x3.svg")
graficos_passos(df_3x3[:2], df_3x3[2:], "img/passos_3x3.svg")
```
